```python
import pandas as pd
import numpy as np
import seaborn as sns
from tensorflow import keras
```

```python
df_train = pd.read_csv('DiseaseTraining.csv')
df_test = pd.read_csv('DiseaseTesting.csv')
```

```python
df_train.head()
```

```python
df_train.drop(columns='Unnamed: 133', inplace=True)
df_train.head(3)
```

```python
df_test.head(3)
```

```python
df_train.info()
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
df_train['prognosis'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df_train['prognosis'] = le.fit_transform(df_train['prognosis'])
df_train.head()
```

```python
df_train['prognosis'].max()
```

```python
df_test['prognosis'] = le.transform(df_test['prognosis'])
df_test.head(3)
```

```python
le.classes_
```

```python
x_train = df_train.drop(columns='prognosis')
y_train = df_train['prognosis'] 
```

```python
x_train.head(2)
```

```python
y_train.head(2)
```

```python
x_test = df_test.drop(columns='prognosis')
y_test = df_test['prognosis']
```

```python
print(y_train[:1])
y_train = keras.utils.to_categorical(y_train, num_classes=41)
print(y_train[:1])
```

```python
print(y_train[:2])
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
x_train.shape
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(x_train.shape[-1],)))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(units=41, activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train, y_train, batch_size=1000, epochs=50, verbose=2)
```

```python
prediction = model.predict(x_test)
print(prediction[:1])
```

```python
prediction = np.argmax(prediction, axis=-1)
print(prediction[:5])
```

```python
print(y_test[:5])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_test, prediction))
```

```python
print(confusion_matrix(y_test, prediction))
```
